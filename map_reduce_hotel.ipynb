Cell 1: Import necessary libraries and load the dataset

```python
import pandas as pd
from collections import defaultdict

# Load the dataset from an existing CSV file
df = pd.read_csv('/content/reviews.csv')
print("Dataset Loaded:")
print(df)
```

Cell 2: Define the mapper function

```python
# Define the mapper function
def mapper(row):
    # Extract the rating from the row
    rating = str(row.rating).strip()  # Ensure rating is a string and strip whitespace
    return (rating, 1)

# Apply the mapper function to each row in the dataset
mapped_results = [mapper(row) for row in df.itertuples(index=False)]
print("\nMapped Results:")
for result in mapped_results:
    print(result)
```

Cell 3: Shuffle and sort

```python
# Shuffle and sort function
def shuffle_sort(mapped_results):
    shuffle_sorted = defaultdict(list)
    for key, value in mapped_results:
        shuffle_sorted[key].append(value)
    return shuffle_sorted

# Shuffle and sort the mapped results
shuffle_sorted_results = shuffle_sort(mapped_results)
print("\nShuffled and Sorted Results:")
for key, values in shuffle_sorted_results.items():
    print(f"{key}: {values}")
```

Cell 4: Define the reducer function

```python

# Define the reducer function
def reducer(key, values):
    return (key, sum(values))

# Apply the reducer function to the shuffled and sorted results
reduced_results = [reducer(key, values) for key, values in shuffle_sorted_results.items()]
print("\nReduced Results:")
for result in reduced_results:
    print(result)
```
